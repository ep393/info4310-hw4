```python
import pandas as pd
import json
```

```python
df = pd.read_csv('alcohol_consumption.csv')
```

```python
df['state'] = df['state'].str.title()
```

```python
all_states_data = {}
```

```python
for state in df['state'].unique():
    state_data = df[df['state'] == state]
    formatted_data = state_data.drop(columns='state').to_dict(orient='records')
    all_states_data[state] = formatted_data
```

```python
all_states_data
```

```python
with open('states-data.json', 'w') as fp:
    json.dump(all_states_data, fp, indent=4)
```
